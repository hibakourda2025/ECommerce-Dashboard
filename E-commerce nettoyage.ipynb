```python
# verification des valeurs manquantes
df.isnull().sum()
df = df.dropna(subset=['Returns'])
```

```python
#verification des doublons
print(df.duplicated().sum())
df = df.drop_duplicates()
# la colonnes 'customer age' et 'age' sont les meme on peut supprimer l'une d'entre elles
df = df.drop(columns=['Customer Age'])
```

```python
#verification des types de données
df.dtypes
```

```python
# changer les types de données
df['Purchase Date'] = pd.to_datetime(df['Purchase Date'])
df['Returns'] = df['Returns'].astype(int)
```

```python
# ajouter des underscores aux noms de colonnes composés de plusieurs mots
df.columns = df.columns.str.replace(' ', '_')
```

```python
# enregistrer le dataframe nettoyé dans un nouveau fichier CSV
df.to_csv('ecommerce_cleaned.csv', index=False)
```
